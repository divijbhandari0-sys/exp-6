```python
#5. given year is a leap year or not
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
#6.write a python program to incriment the date
from datetime import datetime, timedelta

date_str = input("Enter a date (YYYY-MM-DD): ")

try:
    # Parse the input string into a datetime object
    current_date = datetime.strptime(date_str, '%Y-%m-%d')

    # Increment the date by one day
    next_date = current_date + timedelta(days=1)

    print(f"The next day is: {next_date.strftime('%Y-%m-%d')}")
except ValueError:
    print("Invalid date format. Please use YYYY-MM-DD.")
```

```python
'''Problem statement
write a python statement to calculate the gross salary of an employee based

if basic <= 10000: HRA = 20%, DA = 80%

if basic <= 20000: HRA = 25%, DA = 90%

orthewise HRA = 30%, DA = 95%'''

basic_salary = float(input("Enter the basic salary: "))

if basic_salary <= 10000:
    hra = 0.20 * basic_salary
    da = 0.80 * basic_salary
elif basic_salary <= 20000:
    hra = 0.25 * basic_salary
    da = 0.90 * basic_salary
else:
    hra = 0.30 * basic_salary
    da = 0.95 * basic_salary

gross_salary = basic_salary + hra + da

print(f"Basic Salary: {basic_salary}")
print(f"HRA: {hra}")
print(f"DA: {da}")
print(f"Gross Salary: {gross_salary}")
```

```python
''' write a python program to calculate income tax based on annual income

 up to $2,50,000:no tax

 $2,50,001 - $5,00,000: 5%

 $5,00,001 - $10,00,000: 20%

above $10,00,000: 30%
```

```python
annual_income = float(input("Enter your annual income: "))

tax = 0

if annual_income <= 250000:
    tax = 0
elif annual_income <= 500000:
    # 5% on income from $2,50,001 to $5,00,000
    tax = (annual_income - 250000) * 0.05
elif annual_income <= 1000000:
    # Tax on first slab (2,50,001 - 5,00,000) is (500000 - 250000) * 0.05 = 12500
    # Plus 20% on income from $5,00,001 to $10,00,000
    tax = (500000 - 250000) * 0.05  # Tax for the 5% slab
    tax += (annual_income - 500000) * 0.20 # Tax for the 20% slab
else:
    # Tax on first slab (2,50,001 - 5,00,000) is (500000 - 250000) * 0.05 = 12500
    # Tax on second slab (5,00,001 - 10,00,000) is (1000000 - 500000) * 0.20 = 100000
    # Plus 30% on income above $10,00,000
    tax = (500000 - 250000) * 0.05  # Tax for the 5% slab
    tax += (1000000 - 500000) * 0.20 # Tax for the 20% slab
    tax += (annual_income - 1000000) * 0.30 # Tax for the 30% slab

print(f"Annual Income: ${annual_income:,.2f}")
print(f"Income Tax: ${tax:,.2f}")
```
